```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
col1=ColumnTransformer(transformers=[
    ("onehot",OneHotEncoder(drop="first",sparse_output=False),["gender","course","study_method","internet_access"]),
    ("ordinal",OrdinalEncoder(categories=[["poor","average","good"]]),["sleep_quality"]),
    ("faculty_oridianl",OrdinalEncoder(categories=[["low","medium","high"]]),["facility_rating"]),
    ("difficulty_ordinal",OrdinalEncoder(categories=[["hard","moderate","easy"]]),["exam_difficulty"]),
],remainder="passthrough")
```

```python
pipe=Pipeline([
    ("columns",col1),
    ("scaler",StandardScaler()),
    ("model",LinearRegression())
    ])
```

```python
x,y=df.drop("exam_score",axis=1),df["exam_score"]
```

```python
x_tarin,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
pipe.fit(x_tarin,y_train)
```

```python
pred=pipe.predict(x_test)
```

```python
mse=mean_squared_error(y_test,pred)
print("MSE:",mse)

r2=r2_score(y_test,pred)
print("R2:",r2)
```
