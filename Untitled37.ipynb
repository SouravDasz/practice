```python
time_data=pd.to_datetime(df['OrderDate'])
```

```python
year=time_data.dt.year
```

```python
month_name=time_data.dt.month
```

```python
day=time_data.dt.weekday
```

```python
day
```

```python
month=time_data.dt.day
```

```python
month
```

```python
month_day=np.where(month>15,"Month end","Month start")
```

```python
month_day
```

```python
df2 = pd.DataFrame({
    "month_day": month_day,
    "day": day,
    "month": month_name,
    "year": year
})
```

```python
df2.head()
```

```python
df=pd.concat([df,df2],axis=1)
```
