```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
onehot=OneHotEncoder(drop="first",sparse_output=False)
```

```python
city=onehot.fit_transform(df[["city"]])
```

```python
df_part=pd.DataFrame(city,columns=onehot.get_feature_names_out())
df=pd.concat((df,df_part),axis=1)
```

```python
df_part.head()
```

```python
df.head()
```

```python
df=df.drop("city",axis=1)
```

```python
sns.pairplot(df)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
x,y=df.drop("placement",axis=1),df["placement"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
y
```

```python
model1=DecisionTreeClassifier()
model2=LogisticRegression()
model3=SVC()

model1.fit(x_train,y_train)
no1=model1.predict(x_test)
acc=accuracy_score(y_test,no1)
print("dt -->",acc)
print("cross val -->",cross_val_score(model1,x,y,cv=5).mean())

model2.fit(x_train,y_train)
no2=model2.predict(x_test)
acc=accuracy_score(y_test,no2)
print("lr -->",acc)
print("cross val -->",cross_val_score(model2,x,y,cv=5).mean())

model3.fit(x_train,y_train)
no3=model3.predict(x_test)
acc=accuracy_score(y_test,no3)
print("svc -->",acc)
print("cross val -->",cross_val_score(model3,x,y,cv=5).mean())

```

```python
from sklearn.ensemble import VotingClassifier

model4=VotingClassifier(estimators=[("Dt",model1),
                                    ("le",model2)])

model4.fit(x_train,y_train)
```

```python
pred4=model4.predict(x_test)
acc=accuracy_score(y_test,pred4)
print("vc -->",acc)
print("cross val -->",cross_val_score(model4,x,y,cv=5).mean())
```
