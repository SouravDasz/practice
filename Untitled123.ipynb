```python
from gensim.models import Word2Vec
```

```python
word=[
    ["I","love","ml"],
    ["no","can","better","then","me"],
    ["I","am","engineer"]
]
```

```python
model=Word2Vec(
    word,
    vector_size=3,
    window=1,
    sg=1,
    min_count=1
)
```

```python
model.wv["love"]
```

```python
model.wv.most_similar("I")
```

```python
model.wv.get_normed_vectors()
```

```python
import numpy as np
print(model.wv.index_to_key)
print(model.wv.similar_by_vector(np.array([-0.44667298,  0.87598896, -0.18200706], dtype=np.float32)))
```
