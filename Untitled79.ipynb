```python
import keras_tuner as kt
```

```python
def build_model(hp):
  model=Sequential()
  units=hp.Int("units",min_value=20,max_value=200,step=5)
  activaion =hp.Choice("activaion",["relu","tanh","sigmoid"])
  model.add(Input(shape=(df.shape[1],)))
  model.add(Dense(units,activation=activaion))
  model.add(Dense(units,activation=activaion))
  model.add(Dense(1))


  opt=hp.Choice("optimizer",["adam","sgd","rmsprop"])
  model.compile(optimizer=opt,loss="mse",metrics=["mae"])


  return model
```

```python
tuner = kt.RandomSearch(
    build_model,
    objective="val_mae",
    max_trials=5,
    directory="mu_dir",
    project_name="house_prize"
)
```

```python
tuner.search(df,y,epochs=5,validation_split=0.2,batch_size=32)
```

```python
best_hp = tuner.get_best_hyperparameters(num_trials=1)[0]
best_hp.values
```
