```python
from sklearn.preprocessing import LabelEncoder, StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
ohe=OneHotEncoder(sparse_output=False,drop='first')
df2=ohe.fit_transform(df[["Gender","Subscription Type","Contract Length"]])
```

```python
df2=pd.DataFrame(df2,columns=ohe.get_feature_names_out())
```

```python
df=pd.concat([df,df2],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Gender',"Subscription Type","Contract Length"],axis=1,inplace=True)
```

```python
x,y=df.drop("Churn",axis=1),df["Churn"]
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
#
```

```python
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(11,input_dim=13,activation="sigmoid"))
model.add(Dense(5,activation="sigmoid"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy",optimizer="Adam",metrics=["accuracy"])
```

```python
history=model.fit(x_train,y_train,validation_data=(x_test, y_test),epochs=20)
```

```python
history.history
```

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.legend(["train","test"])
plt.show()
```
