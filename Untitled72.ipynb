```python
import numpy as np
import pandas as pd
import keras as keras
from keras import Sequential
from keras.layers import Conv2D,MaxPool2D,BatchNormalization,Dense,Flatten
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
imagen=ImageDataGenerator(rescale=1./255)
```

```python
train_ds=imagen.flow_from_directory(
    directory="/content/training_set/training_set",
    target_size=(256,256),
    batch_size=32,
    class_mode="binary",
    shuffle=True,

)
```

```python
test_ds=imagen.flow_from_directory(
    directory="/content/test_set/test_set",
    target_size=(256,256),
    batch_size=32,
    class_mode="binary",
    shuffle=True,

)
```

```python
import cv2
import matplotlib.pyplot as plt
img,label=test_ds[0]
img=cv2.resize(img[0],(256,256))
plt.imshow(img)
```

```python
model=Sequential([
    Conv2D(32,kernel_size=(3,3),activation="relu",input_shape=(256,256,3)),
    MaxPool2D(pool_size=(2,2)),
    Conv2D(64,kernel_size=(3,3),activation="relu"),
    MaxPool2D(pool_size=(2,2),strides=(2,2)),

    Flatten(),

    Dense(128,activation="relu"),
    Dense(1,activation="sigmoid")


])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
history=model.fit(train_ds,epochs=10,validation_data=test_ds)
```
