```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
nltk.download("stopwords")
def process(text):
  text=re.sub("[^a-zA-Z]"," ",text)
  text=text.lower()
  text=text.split()
  stemmer=PorterStemmer()
  text=[stemmer.stem(word) for word in text if word not in stopwords.words('english')]
  text=' '.join(text)

  return text


  return text
```

```python
process("Waste of money")
```

```python
df["Review_Text"]=df["Review_Text"].apply(process)
```

```python
df.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(max_features=2000)
text_features = count.fit_transform(df["Review_Text"]).toarray()
```

```python
text_df = pd.DataFrame(text_features)
```

```python
df = pd.concat([df.drop("Review_Text", axis=1), text_df], axis=1)
```

```python
df.head(2)
```

```python
onehot=OneHotEncoder(drop="first",sparse_output=False)
part=onehot.fit_transform(df[["State"]])
part=pd.DataFrame(part,columns=onehot.get_feature_names_out())
```

```python
part.isnull().sum()
```

```python
df=pd.concat([df.drop("State",axis=1),part],axis=1)
```

```python
df.head(2)
```

```python
x,y=df.drop("Delivery_Status",axis=1),df["Delivery_Status"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train.columns = x_train.columns.astype(str)
x_test.columns = x_test.columns.astype(str)

scaler=StandardScaler()
scaler_test=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler_test.fit_transform(x_test)

# logistic regression
lg=LogisticRegression(multi_class="multinomial")
lg.fit(x_train,y_train)

lg_pred=lg.predict(x_test)
print("accuracy is --->",accuracy_score(y_test,lg_pred))
print("cross val --->",(cross_val_score(lg,x_train,y_train,cv=5)).mean())
print("confusion matrix -->",confusion_matrix(y_test,lg_pred))

# decissiontree

dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt_pred=dt.predict(x_test)
print("accuracy is --->",accuracy_score(y_test,dt_pred))
print("cross val --->",(cross_val_score(dt,x_train,y_train,cv=5)).mean())
print("confusion matrix -->",confusion_matrix(y_test,dt_pred))

# random forest

rf=RandomForestClassifier()
rf.fit(x_train,y_train)
rf_pred=rf.predict(x_test)
print("accuracy is --->",accuracy_score(y_test,rf_pred))
print("cross val --->",(cross_val_score(rf,x_train,y_train,cv=5)).mean())
print("confusion matrix -->",confusion_matrix(y_test,rf_pred))

# xgb boost

xgb=XGBClassifier()
xgb.fit(x_train,y_train)
xgb_pred=xgb.predict(x_test)
print("accuracy is --->",accuracy_score(y_test,xgb_pred))
print("cross val --->",(cross_val_score(xgb,x_train,y_train,cv=5)).mean())
print("confusion matrix -->",confusion_matrix(y_test,xgb_pred))


# knn

knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
knn_pred=knn.predict(x_test)
print("accuracy is --->",accuracy_score(y_test,knn_pred))
print("cross val --->",(cross_val_score(knn,x_train,y_train,cv=5)).mean())
print("confusion matrix -->",confusion_matrix(y_test,knn_pred))


```
