```python
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
onehot=OneHotEncoder(sparse_output=False,drop="first")
part=onehot.fit_transform(df[["Gender","Department"]])
```

```python
ordinal=OrdinalEncoder(categories=[["Single","Married","Divorced"],["Executive","Analyst","Assistant","Manager"]])

df[["Marital_Status","Job_Role"]]=ordinal.fit_transform(df[["Marital_Status","Job_Role"]])
```

```python
part=pd.DataFrame(part,columns=onehot.get_feature_names_out())
```

```python
part.head()
```

```python
df=pd.concat([df,part],axis=1)
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(["Gender","Department"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df["Overtime"]=df["Overtime"].map({"No":0,"Yes":1})
df["Attrition"]=df["Attrition"].map({"No":0,"Yes":1})
```

```python
x,y=df.drop("Attrition",axis=1),df["Attrition"]
```

```python
x.isnull().sum()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=23,stratify=y)
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
x_train, y_train = sm.fit_resample(x_train, y_train)
```

```python
df=DecisionTreeClassifier(class_weight={0:1,1:3})
df.fit(x_train,y_train)

rf=RandomForestClassifier(class_weight={0:1,1:3})
rf.fit(x_train,y_train)

log=LogisticRegression(class_weight='balanced')
log.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,r2_score,confusion_matrix
df_pred=df.predict(x_test)

print("for decission tree -----")
print(accuracy_score(y_test,df_pred))
print(confusion_matrix(y_test,df_pred))
print()

rf=rf.predict(x_test)

print("for random forest -----")
print(accuracy_score(y_test,rf))
print(confusion_matrix(y_test,rf))
print()

print("for logistic regression -----")
print(accuracy_score(y_test,log.predict(x_test)))
print(confusion_matrix(y_test,log.predict(x_test)))
print()


```
