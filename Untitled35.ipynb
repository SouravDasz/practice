```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
import random
numeric=[]
label=[]

for i in range(100):
  number=random.randint(65,100)
  label_value=1
  numeric.append(number)
  label.append(label_value)
```

```python
for i in range(100):
  number=random.randint(1,70)
  label_value=0
  numeric.append(number)
  label.append(label_value)
```

```python
import numpy as np
import pandas as pd
nparr=np.array(numeric)
nplable=np.array(label)
```

```python
df=pd.DataFrame({"value":nparr,"label":nplable})
```

```python
df.head()
```

```python
model=Sequential([
    Dense(units=16,input_shape=(1,),activation="relu"),
    Dense(units=5,activation="relu"),
    Dense(units=2,activation="softmax")
])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.01),loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
x,y=df["value"].values.reshape(-1,1),df["label"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model.fit(x_train,y_train,epochs=30,batch_size=10,validation_data=(x_test,y_test))
```

```python
num=np.array([[35]])
model.predict(num)
```
