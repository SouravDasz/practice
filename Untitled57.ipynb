```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout, BatchNormalization
from tensorflow.keras.models import Sequential
```

First, let's examine the contents of one of the XML annotation files to understand how the face mask labels are defined. We'll pick one at random, for instance, `annotations/maksssksksss81.xml`.

```python
import xml.etree.ElementTree as ET

xml_file_path = '/content/annotations/maksssksksss81.xml'
tree = ET.parse(xml_file_path)
root = tree.getroot()

print(f"Root tag: {root.tag}")
for child in root:
    print(f"  Child tag: {child.tag}, attributes: {child.attrib}")
    if child.tag == 'object':
        for sub_child in child:
            print(f"    Sub-child tag: {sub_child.tag}, text: {sub_child.text}")
```

```python
import os
import xml.etree.ElementTree as ET
import shutil

annotations_dir = '/content/annotations'
images_dir = '/content/images'
output_base_dir = '/content/processed_dataset'

# Create output directories for classes
class_labels = ['with_mask', 'without_mask', 'mask_weared_incorrect']
for label in class_labels:
    os.makedirs(os.path.join(output_base_dir, label), exist_ok=True)

# Process each XML annotation file
for annotation_file in os.listdir(annotations_dir):
    if annotation_file.endswith('.xml'):
        xml_path = os.path.join(annotations_dir, annotation_file)
        tree = ET.parse(xml_path)
        root = tree.getroot()

        filename = root.find('filename').text
        image_path = os.path.join(images_dir, filename)

        # Initialize a list to store labels for the current image
        image_labels = []

        for obj in root.findall('object'):
            label = obj.find('name').text
            if label in class_labels:
                image_labels.append(label)

        # Determine the primary label for the image for simplicity
        # For now, if multiple masks are present, we'll assign to the first found valid class.
        # A more sophisticated approach might be needed for multi-label images.
        if image_labels:
            primary_label = image_labels[0] # Use the first label found
            destination_dir = os.path.join(output_base_dir, primary_label)
            if os.path.exists(image_path):
                shutil.copy(image_path, destination_dir)
            else:
                print(f"Warning: Image not found for {filename} at {image_path}")

print("Dataset reorganization complete. Images moved to class-specific subdirectories.")
```

```python
train_ds = keras.utils.image_dataset_from_directory(
    directory="/content/processed_dataset",
    labels="inferred",
    label_mode="int",
    batch_size=32,
    image_size=(256,256),
    validation_split=0.2,
    subset="training",
    seed=42
)

val_ds = keras.utils.image_dataset_from_directory(
    directory="/content/processed_dataset",
    labels="inferred",
    label_mode="int",
    batch_size=32,
    image_size=(256,256),
    validation_split=0.2,
    subset="validation",
    seed=42
)
```

```python
def process(image, label):
  image = tf.cast(image/255. , tf.float32)
  return image, label
```

```python
train_ds=train_ds.map(process)
val_ds=val_ds.map(process)
```

```python
model=Sequential()
model.add(Conv2D(30,kernel_size=(3,3),activation="relu",input_shape=(256,256,3)))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(60,kernel_size=(3,3),activation="relu",input_shape=(256,256,3)))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(50,activation="relu"))
model.add(Dense(3,activation="softmax"))
```

```python
from tensorflow.keras.utils import image_dataset_from_directory
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from tensorflow import keras
from tensorflow.keras import layers

data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.1),
])
```

```python
train_ds = train_ds.map(
    lambda x, y: (data_augmentation(x, training=True), y)
)
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss = "sparse_categorical_crossentropy"
,metrics=["accuracy"])
```

```python
histry=model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(histry.history["accuracy"],c="red")
plt.plot(histry.history["val_accuracy"],c="blue",)
plt.legend(["train","val"])
plt.title("model accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
```
